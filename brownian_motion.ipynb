```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
def main():
    n = 10000
    T = 1.
    times = np.linspace(0.,T,n)
    #timestep
    dt = times[1] - times[0]
    #cal values of the brownian motion
    # Bt2 - Bt1~ N(mu = 0,sigma^2 = t2-t1)
    
    #number of simulatenous brownian motions
    
    d = 10000
    
    #increments of brownian motion
    dB = np.sqrt(dt)*np.random.normal(size = (n-1,d))
    #start with some zero and cumulatively add the diffs
    
    B0 = np.zeros(shape=(1,d))
    
    #concatenate for brownian values
    
    B = np.concatenate((B0,np.cumsum(dB,axis = 0)), axis = 0)
    
    #plot
    
    plt.plot(times, B)
    

    
    
    
    
if __name__ == '__main__':
    main()
```
